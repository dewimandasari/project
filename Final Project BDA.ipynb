## Data Preparation

```python
import s3fs
import pandas as pd

s3 = s3fs.S3FileSystem()
s3.ls('dts-fga-bda032')
with s3.open ('s3://dts-fga-bda032/tweetsByDay.csv','rb') as f:
    data = pd.read_csv(f)
    
data.head()
```

Rumusan Masalah :

Rachael Tatman (peneliti) suka mengadakan konferensi secara langsung ketika menghadiri sebuah konferensi, untuk referensi sendiri di kemudian hari dan untuk membantu orang lain yang tidak menghadiri konferensi tetap up-to-date. Setelah konferensi terakhir yang dihadiri, peneliti penasaran: apakah livetweets mendapatkan keterlibatan lebih atau kurang dari jenis tweet lain?

Isi :

Data set di atas berisi informasi dari 313 tweet yang dikirim dari akun twitter pribadi milik Rachael Tatman (peneliti) dari tanggal 29 September 2017 Sampai tanggal 26 Oktober 2017. Penelitian ini bertujuan untukDari setiap tweet yang dikirim, dicatat informasi sebagai berikut :

-at_conference : Apakah tweet tersebut dikirim selama konferensi

-day : Hari pada saat tweet tersebut dikirim

-impressions : Berapa kali tweet tersebut dilihat

-engagements : Berapa kali tweet digunakan (jumlah dari kolom)

-retweets : Berapa kali tweets tersebut diretweet

-likes : Berapa kali tweet disukai

-user profile clicks : Berapa kali seseorang mengklik profile akun peneliti dari tweet

-url clicks : Berapa kali  seseorang yang mengklik URL di tweet (Tidak semua tweet memiliki URL)

-hashtag clicks: Berapa kali seseorang mengklik Hashtag di tweet(Tidak semua tweet memiliki hashtag)

-detail expands: Berapa kali seseorang menyebarluaskan tweet

-follows: Berapa kali seseorang mengikuti akun peneliti dari tweet

-media views: Berapa kali seseorang melihat media yang disematkan dalam tweet (tidak semua tweet memiliki media)

-media engagements: Berapa kali seseorang mengklik media yang disematkan dalam tweet (tidak semua tweet memiliki media)

### </b> 4. Dari 2 features yang dipilih tersebut (No. 3), buatlah analisis model regresi (linear atau nonlinear). Jenis model regresi yang dipilih harus menyesuaikan pola sebaran data dari scatter plot. Model yang telah dibuat harus ditunjukkan secara visual dg grafik, serta tunjukkan tingkat keakuratannya (bisa menggunakan sembarang metrik)<b>

<b> Regresi Linear <b>

```python
msk = np.random.rand(len(dataClean4)) < 0.8
train = dataClean4[msk]
test = dataClean4[~msk] 
```

```python
#persebaran data train
plt.scatter(train.impressions, train.retweets,  color='blue')
plt.xlabel("Impressions")
plt.ylabel("Retweets")
plt.show()
```

```python
# modelling data train use linear Regression 
from sklearn import linear_model
regr = linear_model.LinearRegression()
train_x = np.asanyarray(train[['impressions']])
train_y = np.asanyarray(train[['retweets']])
regr.fit (train_x, train_y)
# The coefficients
print ('Coefficients: ', regr.coef_)
print ('Intercept: ',regr.intercept_)
```

```python
plt.scatter(train.impressions, train.retweets,  color='blue')                    
plt.plot(train_x, regr.coef_[0][0]*train_x + regr.intercept_[0], '-r')
plt.xlabel("Impressions")
plt.ylabel("Retweets")
```

```python
from sklearn.metrics import r2_score

test_x = np.asanyarray(test[['impressions']])
test_y = np.asanyarray(test[['retweets']])
test_y_ = regr.predict(test_x)

print("Model Regresi Linear")
print("Mean absolute error: %.2f" % np.mean(np.absolute(test_y_ - test_y)))
print("Residual sum of squares (MSE): %.2f" % np.mean((test_y_ - test_y) ** 2))
print("R2-score: %.2f" % r2_score(test_y_ , test_y) )
```

```python
x_data, y_data = (dataClean4['impressions'].values, dataClean4['retweets'].values)

# Lets normalize our data
xdata =x_data/max(x_data)
ydata =y_data/max(y_data)
```

```python
def sigmoid(x, Beta_1, Beta_2):
     y = 1 / (1 + np.exp(-Beta_1*(x-Beta_2)))
     return y

def eksponensial(x, Beta_0, Beta_1):
    y = Beta_0*np.exp(Beta_1*x)
    return y
```

```python
from scipy.optimize import curve_fit
popt1, pcov1 = curve_fit(sigmoid, xdata, ydata, maxfev=20000)
popt2, pcov2 = curve_fit(eksponensial, xdata, ydata, maxfev=20000)
```

```python
x = np.linspace(10, 200, 1000)
x = x/max(x)
plt.figure(figsize=(8,5))
y1 = sigmoid(x, *popt1)
y2 = eksponensial(x, *popt2)
plt.plot(xdata, ydata, 'ro', label='data')
plt.plot(x,y1, linewidth=3.0, label='Sigmoid')
plt.plot(x,y2, linewidth=3.0, label='Eksponensial')
plt.legend(loc='best')
plt.xlabel("Impressions")
plt.ylabel("Retweets")    
plt.show()
```

```python
msk = np.random.rand(len(dataClean4)) < 0.8
train_x = xdata[msk]
test_x = xdata[~msk]
train_y = ydata[msk]
test_y = ydata[~msk]

# build the model using t rain set
popt1, pcov = curve_fit(sigmoid, train_x, train_y,maxfev=20000)
popt2, pcov = curve_fit(eksponensial, train_x, train_y,maxfev=20000)

# predict using test set
y_hat1 = sigmoid(test_x, *popt1)
y_hat2 = eksponensial(test_x, *popt2)

# evaluation
print("Model Sigmoid")
print("Mean absolute error Sigmoid: %.2f" % np.mean(np.absolute(y_hat1 - test_y)))
print("Residual sum of squares (MSE) Sigmoid: %.2f" % np.mean((y_hat1 - test_y) ** 2))
from sklearn.metrics import r2_score
print("R2-score Sigmoid: %.2f" % r2_score(y_hat1 , test_y) )

print('Model Eksponensial')
print("Mean absolute error Eksponensial: %.2f" % np.mean(np.absolute(y_hat2 - test_y)))                                                                              
print("Residual sum of squares (MSE) Eksponensial: %.2f" % np.mean((y_hat2 - test_y) ** 2))
print("R2-score Eksponensial: %.2f" % r2_score(y_hat2 , test_y) )
```
